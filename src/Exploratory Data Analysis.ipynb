```python
import pandas as pd
import re
import csv
```

```python

with open('static/coins.txt') as f:
    reader = csv.reader(f)
    next(reader)
    for_df = []
    for row in reader:
        # This file is a tricky - couldn't think of a clever way with regex, so here we are. 
        _row = row[0].strip()
        name = _row[:18].strip()
        cap = _row[18:33].strip()
        price = _row[39:55].strip()
        supply = _row[57:75].strip()
        status = _row[78:].strip()
        for_df.append({
            "Name": name,
            "Market Cap": cap,
            "Price": price,
            "Supply": supply,
            "Status": status
        })


```
